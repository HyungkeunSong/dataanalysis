## 연산자끼워넣기

과제링크:

```python
import numpy as np
import itertools
from itertools import permutations
from functools import reduce
import sys
import time
import operator
import math
```

```python
def q_calculator_old(n, numbers, arithmetics):
    ops_dict = {0: operator.add, 1: operator.sub, 2: operator.mul, 3: lambda x, y: divmod(x,y)[0]}
    ops_list = [v[0] for v in list(enumerate(arithmetics)) for _ in range(0,v[1])]
    tot_ops = list(set(permutations(ops_list)))  
    tot_list = list(permutations(numbers))
    result_list = list()
    
    for elem_list in tot_list:
        for i, v in enumerate(elem_list):
            if i == 0:
                result = v
            else:
                for ops in tot_ops:
                    for op in ops:
                        result = ops_dict[op](result, v)
            
            result_list.append(result)
    return max(result_list), min(result_list)
```

```python
def q_calculator_new(n, numbers, arithmetics):
    ops_dict = {0: operator.add, 1: operator.sub, 2: operator.mul, 3: lambda x, y: math.trunc(x/y)}
    ops_list = [v[0] for v in list(enumerate(arithmetics)) for _ in range(0,v[1])]
    tot_ops = list(set(permutations(ops_list)))  
    result_list = list()
    
    for ops in tot_ops:
        for i, v in enumerate(numbers):
            if i == 0:
                result = v
            else:
                result = ops_dict[ops[i-1]](result, v)
        result_list.append(result)

    return max(result_list), min(result_list)
```

```python
n = 6
numbers = [1,2,3,4,5,6]
arithmetics = [2,1,1,1]
q_calculator_new(n, numbers, arithmetics)
```

```python
n = 2
numbers = [5,6]
arithmetics = [0,0,1,0]
q_calculator_new(n, numbers, arithmetics)
```

```python
n = 3
numbers = [3,4,5]
arithmetics = [1,0,1,0]
q_calculator_new(n, numbers, arithmetics)
```
